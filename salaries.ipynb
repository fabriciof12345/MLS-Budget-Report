```
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plt

url = requests.get("https://mlsplayers.org/resources/salary-guide")
df_list = pd.read_html(url.text)
df = df_list[0]
df.head()
```

```
df['Base Salary'] = df['Base Salary'].str.replace('$','')
df['Base Salary'] = df['Base Salary'].str.replace(',','')
df['Guaranteed Compensation'] = df['Guaranteed Compensation'].str.replace('$','')
df['Guaranteed Compensation'] = df['Guaranteed Compensation'].str.replace(',','')
convert_dict = {
                "Base Salary" : float,
                "Guaranteed Compensation" : float
               } 

df = df.astype(convert_dict)
```

```
df = df[df['Base Salary'] < 500000.00]
df.sort_values(by='Base Salary', ascending=False,inplace=True)
df.head()
```

```
def containsDefGK(position):
  if position.find("D") or position.find("D-M"):
    return True
  else:
    return False
```

```
#This will return all defenders in the league with a base salary under $500,000
df = df[(df['Position(s)'] == "D")]
```
